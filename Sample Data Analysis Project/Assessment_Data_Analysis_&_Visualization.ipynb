## Data Analysis and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read data into pandas dataframe:
cleaned_client_data_df = pd.read_excel(r'C:\Users\FashinLuk\Downloads\Assessment_file\cleaned_assessment_datasets.xlsx')
```

### Create a chart that displays the average client income trend over time

```python
# Calculate the average income per year
average_income_by_year = cleaned_client_data_df.groupby('Year')['Income'].mean().reset_index()

# # Plotting a line chart of the average client income over time
plt.figure(figsize=(12, 8))
plt.plot(average_income_by_year['Year'], average_income_by_year['Income'], marker='o', linestyle='-', color='magenta')
plt.xlabel('Year')
plt.ylabel('Average Income ($)')
plt.title('Average Client Income by Year')
plt.grid(True)
plt.tight_layout()

# Save the plots as a PNG file
plt.savefig('average_income_by_year.png')

# Display plot
plt.show()

# Display the table of average incomes
print(round(average_income_by_year), 0)
```

### Create a chart that displays the average client income by age

```python
# Calculate the average income by age
average_income_by_age = cleaned_client_data_df.groupby('Age')['Income'].mean().reset_index()

# Plotting a line chart of the average income by age of client
plt.figure(figsize=(10, 6))
plt.plot(average_income_by_age['Age'], average_income_by_age['Income'], marker='o', linestyle='-', color='green')
plt.xlabel('Age')
plt.ylabel('Average Income ($)')
plt.title('Average Client Income by Age')
plt.grid(True)
plt.tight_layout()

# Save the plots as a PNG file
plt.savefig('average_income_by_age.png')

# Display plot
plt.show()

# Display the table of average incomes
print(round(average_income_by_age), 0)
```

### Create a chart that displays the average client income by location (city)

```python
# Calculate the average income by city
average_income_by_city = cleaned_client_data_df.groupby('City')['Income'].mean().reset_index()

# Plotting the average income across cities bar chart
plt.figure(figsize=(12, 8))
plt.bar(average_income_by_city['City'], average_income_by_city['Income'], color='lightblue')
plt.xlabel('City')
plt.ylabel('Average Income ($)')
plt.title('Average Client Income by City')
plt.xticks(rotation=45)
plt.tight_layout()

# Save the plots as a PNG file
plt.savefig('average_income_by_city.png')


# Display plot
plt.show()

# Display the table of average incomes
print(round(average_income_by_city), 0)
```
